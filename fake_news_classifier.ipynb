```python
import csv
import pandas as pd

def load_bad_csv(path):
    rows = []
    with open(path, "r", encoding="utf-8", errors="replace") as f:
        reader = csv.reader(f)
        for row in reader:
            rows.append(row)
    return pd.DataFrame(rows[1:], columns=rows[0])


data_fake = load_bad_csv("Fake.csv")
data_true = load_bad_csv("True.csv")
```

```python
import re
import string

def wordopt(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)          # remove urls
    text = re.sub(r'<.*?>+', '', text)                         # remove html tags
    text = re.sub(r'$$   .*?   $$', '', text)                        # remove [content]
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)                   # keep only letters + space
    text = re.sub(r'\s+', ' ', text)                           # multiple spaces → one
    text = text.strip()
    return text
```

```python
import re
import string

data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer(max_features=50000, ngram_range=(1,2), min_df=3, max_df=0.9, stop_words='english')
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
print (classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)

pred_dt = DT.predict(xv_test)

DT.score(xv_test, y_test)
```

```python
print (classification_report(y_test, pred_lr))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)

pred_gb = GB.predict(xv_test)

GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    # pred_RF = RF.predict(new_xv_test)

    return print("\n\nLR Predicition: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction:{}".format(output_lable(pred_LR[0]),
                                                                                                             output_lable(pred_DT[0]),
                                                                                                             output_lable(pred_GB[0])))
```

```python
vectorization = TfidfVectorizer(
    ngram_range=(1,2),
    max_df=0.9,
    min_df=5,
    max_features=50000
)

xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import f1_score, classification_report

svm_model = LinearSVC()
svm_model.fit(xv_train, y_train)

pred_svm = svm_model.predict(xv_test)

print(classification_report(y_test, pred_svm))
f1_new = f1_score(y_test, pred_svm)
f1_new
```
